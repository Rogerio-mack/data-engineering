# Web APIs

APIs, ou Interfaces de Programação de Aplicações, são pontes que conectam sistemas, permitindo que eles se comuniquem entre si, mesmo quando são desenvolvidos em diferentes plataformas ou linguagens. Esse *desacoplamento* é uma das grandes vantagens do uso de APIs e que as torna bastante úteis tanto para obtenção de dados como para expor dados gerados por uma aplicação.


<img src="https://github.com/Rogerio-mack/data-engineering/blob/main/figs/Lubanovic_FastAPI.png?raw=true" height="250" align="left"/></a>


Existem diferentes tipos de APIs.  No contexto do desenvolvimento moderno e de serviços web (**Web Services**), as APIs SOAP e RESTful são amplamente utilizadas por sua simplicidade e robustez.

O **RESTful APIs (Representational State Transfer)** tem uma arquitetura que se baseia em princípios como **statelessness**  e representações de recursos via URLs. Ela é mais flexível que  o SOAP, permite que as aplicações troquem dados em vários formatos e é bastante escalável. Normalmente, os dados são transferidos em JSON, mas também podem ser em XML, HTML ou outros formatos. Por exemplo o GitHub expõe APIs nesse formato. O REST usa o **protocolo HTTP** para transferir dados

O **SOAP (Simple Object Access Protocol)** é altamente estruturada, rígido e usa o formato de dados XML. Ele funciona sobre diversos protocolos, como HTTP, SMTP, TCP, etc. em que os *pacotes* SOAP criados são transmitidos. Devido essa estrutura mais rígida é um protocolo mais seguro e acaba sendo o protocolo preferencial de aplicações comerciais como meios de pagamentos e vendas.

Existem outros tipos de APIs, como APIs locais, RPC (Remote Procedure Call), GraphQL (alternativa ao RESTfull da Meta) etc. mas para os nossos propósitos nos concentraremos nas APIs REST.

No contexto do desenvolvimento moderno, as APIs RESTful são amplamente utilizadas por sua simplicidade e robustez.




# APIs REST

As APIs REST operam através de **endpoints**, isto é, URLs específicas que representam diferentes funções ou dados. Um conceito importante em APIs RESTful é o uso de métodos HTTP **(GET, POST, PUT, DELETE)** para indicar a operação que queremos realizar em determinado recurso. É uma convenção que, embora permita fazer um *PUT em um GET* recomenda-se seguir essa convenção por uma boa prática.

O formato mais comun utilizado pelas APIs REST é o **JSON** (JavaScript Object Notation), que é um formato leve de intercâmbio de dados e independente de plataforma. É facilmente legível (pelo ser humano) e também simples de processar.

O uso de ferramentas como o requests em Python para fazer chamadas HTTP é bastante intuitivo e é um ótimo ponto de partida para você entender o uso de APIs no consumo de dados.

# Wrappers

Wrapper é uma função ou biblioteca que tem por objetivo  chamar uma segunda sub-rotina com pouca ou nenhuma computação adicional. Ele *envelopa* uma funcionalidade de forma a tornar mais amigável a operação, embutindo muitas vezes vários passos da operação. Como o wrapper do Spotify que você viu antes, existem muitos outro que podem facilitar o uso de uma API, como por exemplo o **`PySUS`**.

# Criando sua API

Para criar uma API RESTful você vai precisar de um servidor Web e, em geral, um framework para criar aplicações que executam nesse servidor. Existem vários frameworks populares (vamos nos restringir ao Python) como o **Flask, Django, Tornado, Pyramid e o FastAPI**. Cada tem suas vantagens e desvantagens, das opções mais simples e minimalistas do Flask e FastAPI, às opções mais robustas como o Django e Pyramid, e a escolha dependerá das necessidades específicas do seu projeto. Para nossos propósitos empregaremos o FastAPI.



# FastAPI

O **FastAPI** é um framework moderno, rápido (high-performance) para construir APIs em Python, baseado em Starlette e Pydantic. Ele utiliza tipagem do Python para definir esquemas de validação de dados e oferece suporte automático à geração de documentação interativa (Swagger UI e ReDoc).
